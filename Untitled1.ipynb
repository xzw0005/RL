```python
import random
import gym
import numpy as np
import matplotlib.pyplot as plt

from collections import deque
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras import backend as K
```

```python
TRAIN_EPISODES = 5000
TEST_EPISODES = 100

MEMORY_SIZE = 15000
BATCH_SIZE = 32
GAMMA = .99
LEARNING_RATE = .01 #.001 #.00025 

MAX_EPSILON = 1.0
MIN_EPSILON = .01
EXPLORATION_DURATION = TRAIN_EPISODES / 5
# DECAY_RATE = -np.log(1e-4)/EXPLORATION_STEPS
DECAY_RATE = .985

UPDATE_TARGET = 10000
```

```python
def _huber_loss(target, prediction):
    # sqrt(1+error^2)-1
    error = prediction - target
    return K.mean(K.sqrt(1+K.square(error))-1, axis=-1)
```

```python
class Agent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = LEARNING_RATE
        self.memory = deque(maxlen=MEMORY_SIZE)
        self.steps = 0
        self.epsilon = MAX_EPSILON
        self.model = self._build_model()
        self.target_model = self._build_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        model = Sequential()
        model.add(Dense(50, input_dim=self.state_size, activation='relu'))
        model.add(Dense(30, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
#         model.compile(loss=_huber_loss, optimizer=Adam(lr=self.learning_rate))
        model.compile(optimizer=Adam(lr=self.learning_rate), loss='mse')
        return model

    def update_lr(self):
        if self.learning_rate > 2.5e-4:
            self.learning_rate *= .5
            self.model.optimizer.lr.assign(self.learning_rate)

    def update_target_model(self):
        # copy weights from model to target_model
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def action(self, state, epsilon=None):
        if epsilon is None:
            epsilon = self.epsilon
        if np.random.random() <= epsilon:
            return np.random.randint(self.action_size)
        act_values = self.model.predict(state.reshape(1, state.shape[0]))[0]
        return np.argmax(act_values)  # returns action

    def replay(self, batch_size, train_step):
        minibatch = np.array(random.sample(self.memory, batch_size))
        state_batch = np.array(minibatch[:,0].tolist())
        action_batch = np.array(minibatch[:,1].tolist())
        rewards_batch = np.array(minibatch[:,2].tolist())
        next_state_batch = np.array(minibatch[:,3].tolist())
        is_terminal_batch = np.array(minibatch[:,4].tolist())
        state_value_batch = self.model.predict_on_batch(state_batch)
#         next_state_value_batch = self.target_model.predict_on_batch(next_state_batch)
        next_state_value_batch = self.model.predict(next_state_batch)
        for i in range(len(minibatch)):
            if is_terminal_batch[i]:
                state_value_batch[i, action_batch[i]] = rewards_batch[i]
            else:
                state_value_batch[i, action_batch[i]] = rewards_batch[i] + GAMMA * np.max(next_state_value_batch[i])
        self.model.train_on_batch(state_batch, state_value_batch)
        
    def load(self, name):
        self.target_model.load_weights(name)

    def save(self, name):
        self.target_model.save_weights(name)
```

```python
env = gym.make('CartPole-v1')
# env = gym.make('LunarLander-v2')
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = Agent(state_size, action_size)
```

```python
###### Pretrain ######
while len(agent.memory) < BATCH_SIZE*1: # MEMORY_SIZE:
    state = env.reset()
#     state = np.reshape(state, [1, state_size])
    while True:
        action = random.randrange(action_size)
        next_state, reward, done, _ = env.step(action)
#         next_state = np.reshape(next_state, [1, state_size])
        agent.remember(state, action, reward, next_state, done)
        state = next_state
        if done:
            break
```

```python
train_step = 0
total_rewards_list = []
for ep in range(TRAIN_EPISODES):
    state = env.reset()
    total_rewards = 0
    time = 0
    while True: 
        time += 1
        train_step += 1
        if (ep % 100 == 0):
            env.render()
        action = agent.action(state)
        next_state, reward, done, _ = env.step(action)
        total_rewards += reward
        agent.remember(state, action, reward, next_state, done)
        state = next_state
#         if total_rewards > 200:
#             agent.save('CartPoleModel')
            
        if done: 
            total_rewards_list.append((ep, total_rewards))
            print("episode: {}/{}, time steps: {}, total reward: {:.4}, epsilon: {:.2}"
                  .format(ep, TRAIN_EPISODES, time, total_rewards, agent.epsilon))
            break
    agent.replay(BATCH_SIZE, train_step)
    if agent.epsilon > MIN_EPSILON:
#         agent.epsilon = max(MAX_EPSILON - (MAX_EPSILON - MIN_EPSILON) * ep / EXPLORATION_DURATION, MIN_EPSILON)
        agent.epsilon = max(agent.epsilon * DECAY_RATE, MIN_EPSILON)
```

```python
def running_mean(x, n):
    cumsum = np.cumsum(np.insert(x, 0, 0))
    return (cumsum[n:] - cumsum[:-n]) / n 

eps, rewards = np.array(total_rewards_list[:-TEST_EPISODES]).T
smoothed_rewards = running_mean(rewards, 100)  
plt.plot(eps[-len(smoothed_rewards):], smoothed_rewards) 
plt.plot(eps, rewards, color='gray', alpha=.3)
plt.xlabel('Training Episode')
plt.ylabel('Total Reward')
plt.show()            

env.close()
```
